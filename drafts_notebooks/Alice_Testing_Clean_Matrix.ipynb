```python
def img_plot(img):
    
    fig, axes = plt.subplots(1, 4, figsize = (20, 10))

    max_lum = img.max()

    for i in range(4):
        axes[i].imshow(img[:, :, i], cmap = 'gray', vmin = 0, vmax = max_lum)
```

```python
img_plot(X_sample)
```
