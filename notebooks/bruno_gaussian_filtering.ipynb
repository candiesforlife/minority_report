# 2D Gaussian convolution

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.ndimage import gaussian_filter
```

```python
img = np.zeros((30,30))
img[15,15] = 1
plt.imshow(img, cmap="gray")
```

```python
# Convolution a dilution isotrope
img_conv = gaussian_filter(img, sigma=(1,1))
plt.imshow(img_conv, cmap='gray')
```

## 3D Gaussian convolution

```python
#Let's build a 3D array of axis: (t,x,y)
img3D = np.zeros((10,10,10))
img3D[5,5,5] = 1

img3D_conv = gaussian_filter(img3D, sigma=(2,2,2))

max_lum = img3D_conv.max()

for i in range(9):
    plt.imshow(img3D_conv[i+1,:,:], cmap='gray', vmin=0, vmax=max_lum)
    plt.show()
```
