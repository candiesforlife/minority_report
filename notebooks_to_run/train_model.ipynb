```python
with open(X_train_path, 'rb') as f:
    X_train = pickle.load(f)
with open(X_test_path, 'rb') as f:
    X_test = pickle.load(f)
with open(y_train_path, 'rb') as f:
    y_train = pickle.load(f)
with open(y_test_path, 'rb') as f:
    y_test = pickle.load(f)
```

```python
# Reshape pickles to fit it into the model
X_train = X_train.reshape(-1, X_train.shape[1], X_train.shape[2], X_train.shape[3],1)
X_test = X_test.reshape(-1,  X_test.shape[1], X_test.shape[2], X_test.shape[3],1)
y_train = y_train.reshape(-1, y_train.shape[1], y_train.shape[2], y_train.shape[3],1)
y_test = y_test.reshape(-1, y_test.shape[1], y_test.shape[2], y_test.shape[3],1)
```

```python
def init_model():
    print('init model')
    model = models.Sequential()
  
    print('3D conv 1')
    model.add(layers.Conv3D(32, kernel_size = (4,4,4), activation = 'relu', padding='same',
                                input_shape = (192, 132, 16,1)))
    model.add(layers.MaxPooling3D(2))
  
    print('3D conv 2')
    model.add(layers.Conv3D(128, kernel_size = (3,3,3), activation = 'relu', padding='same'))
    model.add(layers.MaxPooling3D(2))
  
    print('3D conv 3')
    model.add(layers.Conv3D(64, kernel_size = (2,2,2), activation = 'relu', padding='same'))
    model.add(layers.MaxPooling3D(2))
  
    print('3D conv 4')
    model.add(layers.Conv3D(16, kernel_size = (2,2,2), activation = 'relu', padding='same'))
    model.add(layers.MaxPooling3D(2))
    
    print('Flatten and Reshape')
    model.add(layers.Flatten())
    model.add(layers.Dense(1584))
    model.add(layers.Reshape((48,33,1,1)))
    
    print('3D conv 2')
    model.add(layers.UpSampling3D(size=(2, 2, 2)))
    model.add(layers.Conv3D(64, kernel_size=(3,3,3), activation='relu', padding='same'))
  
    print('3D conv 3')
    model.add(layers.UpSampling3D(size=(2, 2, 2)))
    model.add(layers.Conv3D(1, kernel_size=(3,3,3), activation='relu', padding='same'))
  
  
    optimizer = Adam(learning_rate = 0.0001)
  
    model.compile(loss ='mse',
               optimizer=optimizer,
               metrics='mae')
    return model
```

```python
model = init_model()
model.summary()
```

## Run model

```python
es = EarlyStopping(patience=3, restore_best_weights=True)

history = model.fit(X_train, y_train, validation_split=0.3, epochs=200, batch_size=4, callbacks=[es])
```

```python
plt.plot(history.history['val_loss'])
```

```python
plt.plot(history.history['val_mae'])
```

```python
model.evaluate(X_test, y_test)
```
